# PCA for dimension reduction

```python
#Data-Preprocessing: Standardizing the data
from sklearn.preprocessing import StandardScaler
std_data=StandardScaler().fit_transform(df)
print(std_data.shape)
```

```python
#finding the co-variance matrix
sample_data=std_data
covar_matrix=np.matmul(sample_data.T, sample_data)
covar_matrix.shape
```

```python
#finding the top two eigen_values and corresponding eigen-vector as for projecting onto a 2-dim space
from scipy.linalg import eigh
#eigh value generates values and vectors in ascending order so here we are interested only in top two values
e_values,e_vectors=eigh(covar_matrix,eigvals=(48,49))
print(e_vectors.shape)
#converting the shape for further 
e_vectors=e_vectors.T
print(e_vectors.shape)
```

```python
#projecting the original data sample on the plane formed by two principal eigen vectors
new_coordinates=np.matmul(e_vectors,sample_data.T)
print(new_coordinates.shape)
```

```python
PCA_label=data['machine_status']
```

```python
#appending label to the 2d projected data
new_coordinates=np.vstack((new_coordinates,PCA_label)).T
df=pd.DataFrame(data=new_coordinates,columns=("first_principal","second_principal","label"))
print(df.head())
```

```python
sns.FacetGrid(df, hue="label", size=6).map(plt.scatter, 'first_principal', 'second_principal').add_legend()
plt.show()
```

# PCA using scikit learn

```python
from sklearn.decomposition import PCA
from sklearn import decomposition
pca=decomposition.PCA()
pca.n_components=49
pac_data=pca.fit_transform(df)
per_var_ex=pca.explained_variance_ / np.sum(pca.explained_variance_)
cum_var_ex= np.cumsum(per_var_ex)

plt.figure(1, figsize=(6,4))
plt.clf()
plt.plot(cum_var_ex,linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('cum_percentage_explained_variance')
plt.show()
```

```python
pcaa = PCA(n_components=2)
data_reduce = pcaa.fit_transform(df)
print(pcaa.explained_variance_ratio_)
```

# Kmean for clustering

Kmean on dimensionally reduced dataset

```python
data2d=pd.DataFrame({'sensorx': data_reduce[:, 0], 'sensory': data_reduce[:, 1]})
from sklearn.cluster import KMeans
data2d
```

```python
k=3
kmeans=KMeans(n_clusters=k)
kmeans.fit(data2d)
```

```python
labels=kmeans.predict(data2d)
centroids=kmeans.cluster_centers_
```

```python
fig=plt.figure(figsize=(5,5))
colmap={1:'r', 2:'g', 3:'b'}
colors=map(lambda x: colmap[x+1], labels)
colors1=list(colors)
plt.scatter(data2d['sensorx'], data2d['sensory'], color=colors1, alpha=0.5, edgecolor='k')

for idx,centroid in enumerate(centroids): 
    plt.scatter(*centroid, color=colmap[idx+1])
plt.show()

```
